## **Top categories on Google Playstore?**

```
play_store_not_null.groupby("Category")["App"].count().sort_values(ascending= False)
```

```
y = play_store_not_null['Category'].value_counts().index
x = play_store_not_null['Category'].value_counts()
xsis = []
ysis = []
for i in range(len(x)):
    xsis.append(x[i])
    ysis.append(y[i])
```

```
plt.figure(figsize=(18,8))
plt.xlabel("Counts")
plt.ylabel("Categories")

graph = sns.barplot(x = xsis, y = ysis, palette= "husl")
graph.set_title("Top categories on Playstore", fontsize = 25);
```

# **#Basic pie chart to view distribution of apps across various categories**

```
fig, ax = plt.subplots(figsize=(20, 20), subplot_kw=dict(aspect="equal"))
number_of_apps = play_store_not_null["Category"].value_counts()
labels = number_of_apps.index
sizes = number_of_apps.values
ax.pie(sizes,labeldistance=2,autopct='%1.1f%%')
ax.legend(labels=labels,loc="right",bbox_to_anchor=(0.9, 0, 0.5, 1))
ax.axis("equal")
plt.show()
```

## **Number of Apps per Category**

```
plt.figure(figsize=(15,5))
ax = sns.countplot(x='Category', data=play_store_not_null, palette="Set1", order = play_store_not_null['Category'].value_counts().index)
plt.ylabel("Number of Apps")
plt.title("Number of Apps per Category", size=20)
for item in ax.get_xticklabels():
    item.set_rotation(90)
```

## **Application categories with their average rating**

Mean and standard deviation plot

```
byCategory = play_store_not_null.groupby('Category')
plt.figure(figsize=(35,15))
std = byCategory.std()
std = pd.DataFrame(std[['Rating']])
avg = byCategory.mean()
avg = pd.DataFrame(avg[['Rating']])

chart5 = avg.plot(figsize=(15,5),legend=False,kind="bar",rot=90,fontsize=12,color="blue")
chart5.set_title("Application categories with their average rating", fontsize=20)
chart5.set_xlabel("Application Classification", fontsize=20)
chart5.set_ylabel("App Rating", fontsize=20)
chart5.set_ylim(0,5)
fig_5 = chart5.get_figure()
fig_5.savefig("output.png", dpi=300, bbox_inches = "tight")
```

## **Analysis after Merginig two table**


```
merged_frame = play_store.merge(user_review, on="App")
```

```
merged_frame.head()
```

```
msno.matrix(merged_frame)
```

```
category_sentiment = merged_frame.groupby(['Category','Sentiment']).size().reset_index(name='Sentiment Count')
```

```
category_sentiment['log_sentiment_count'] = np.log2(category_sentiment['Sentiment Count'])
```

## **Number of installs type wise according to Genres**

```
plt.figure(figsize=(10,7))
plt.xticks(rotation=90,fontsize=11)
plt.xlabel("Category", fontsize=15)
plt.ylabel("Installs", fontsize=15)
plt.title("Number of installs type wise according to Genres",fontsize=10)
sns.barplot('Category', 'log_sentiment_count', hue='Sentiment', data=category_sentiment);
```
